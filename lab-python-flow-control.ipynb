```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print (products) 

inventory = {}
for product in products:
    quantity = int(input(f"Enter the quantity for {product}: "))
    inventory[product] = quantity

print(inventory)  


```

```python
customer_orders = set()
while True:
    product = input("Enter the name of a product to order: ")
    customer_orders.add(product)
    another = input("Do you want to order another product? (yes/no): ")
    if another.lower() != "yes":        
        break   
    

print(customer_orders) 
```

```python
total_products_ordered = len(customer_orders)
percentage_orders = (total_products_ordered / len(products)) * 100
order_status = (total_products_ordered, percentage_orders)

print("order statistics:")
print(f"Total products ordered: {order_status[0]}")
print(f"Percentage of products ordered: {order_status[1]}%") 
```

```python
for product in customer_orders:
    inventory[product] -= 1 
```

```python
print("Updated inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}") 
```
